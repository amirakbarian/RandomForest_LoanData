Random Forest Classification on Loan data from LendingClub.com
The CSV file has been already cleaned 

```python
# Importing the required libraries:
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading the csv file:
loan = pd.read_csv('loan_data.csv')
```

```python
# the first 5 rows of the loan dataframe:
loan.head(5)
```

```python
# checking the data types, number of rows and whether there are null values
loan.info()
```

```python
loan.describe()
```

### some visualizations to better understand the data:

```python
# Histogram of Fico distributions for each credit.policy column:
plt.figure(figsize=(12,5))
sns.distplot(loan[loan['credit.policy']==0]['fico'], axlabel='FICO', kde=False, label="credit policy = 0")
sns.distplot(loan[loan['credit.policy']==1]['fico'], axlabel='FICO', kde=False, label="credit policy = 1")
plt.legend()
```

```python
# Histogram of Fico distributions for each not.fully.paid column:
plt.figure(figsize=(12,5))
sns.distplot(loan[loan['not.fully.paid']==0]['fico'], axlabel='FICO', kde=False, label="not fully paid = 0")
sns.distplot(loan[loan['not.fully.paid']==1]['fico'], axlabel='FICO', kde=False, label="not fully paid = 1")
plt.legend()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='purpose',hue='not.fully.paid',data=loan)
```

```python
plt.figure(figsize=(14,14))
sns.jointplot(x='fico',y='int.rate',data=loan)
```

```python
sns.lmplot(data=loan,x="fico",y="int.rate", col="not.fully.paid",hue='credit.policy')
```
